```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Load MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Randomly pick 20 images
random_indices = np.random.choice(train_images.shape[0], 20, replace=False)
batch_images = train_images[random_indices]

# Plot the batch
plt.figure(figsize=(10, 5))
for i in range(20):
    plt.subplot(4, 5, i + 1)  # Create a 2x10 grid for images
    plt.imshow(batch_images[i], cmap='grey')
    plt.axis('off')
plt.show()
```

```python
# Pick one image
image = train_images[0]

# Function to visualize shifts
def shift_window(image, window_size=18):
    height, width = image.shape
    for i in range(0, height - window_size + 1):
        for j in range(0, width - window_size + 1):
            # Extract window
            window = image[i:i + window_size, j:j + window_size]
            plt.imshow(window, cmap='gray')
            plt.axis('off')
            plt.show()

# Visualize shifts
shift_window(image)
```

```python
# Reshape one image to (56, 14)
reshaped_image = train_images[0].reshape(56, 14)
plt.imshow(reshaped_image, cmap='gray')
plt.title("Reshaped to 56x14")
plt.axis('off')
plt.show()


#Reshape to (56, 14):
#The image's pixel grid is "stretched" to fit the new dimensions.
#The rows are split into two parts and aligned into a new structure.
#The visual result is a distorted version of the original.
```

```python
# Reshape one image to (784, 1)
reshaped_image = train_images[0].reshape(784, 1)
plt.imshow(reshaped_image, cmap='gray', aspect=0.05)  # Adjust aspect ratio for visualization
plt.title("Reshaped to (784, 1)")
plt.axis('off')
plt.show()

#Reshape to (784, 1):
#The 2D image is flattened into a single column of 784 pixels.
#The image becomes a vertical line of pixel intensity values.
#This representation is common for ML models that treat images as 1D vectors.
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```
